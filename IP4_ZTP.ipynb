### 4. Plotting Price doesn't work because it has the object data type not integer. 

```python
car_sales['Price']=car_sales['Price'].str.replace('$', '')
```

```python
car_sales['Price']=car_sales['Price'].str.replace(',', '')
```

```python
car_sales.head()
```

```python
car_sales=car_sales.astype({'Price': 'int64'})
```

```python
car_sales.info()
```

```python
car_sales['Total Sales']=np.cumsum(car_sales['Price'])
```

```python
car_sales.head()
```

```python
car_sales['Sale Date']=pd.date_range(start='2-12-23', periods=len(car_sales))
car_sales.head()
```

```python
plt.plot(car_sales['Price'], 
         car_sales['Total Sales'])
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Price vs Total Sales Line Plot')
plt.show()
```

```python
plt.scatter(car_sales['Odometer (KM)'], 
            car_sales['Price'])
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Odometer vs Price Scatter Plot')
```

```python
plt.bar(car_sales['Make'], 
            car_sales['Odometer (KM)'])
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.title('Make vs Odometer Bar PLot')
```

```python
plt.hist(car_sales['Make'])
plt.xlabel('Make')
plt.title('Histogram of Car Make')
```

```python
fig=plt.figure(figsize=(8,8))
fig.suptitle('Car Sales Plots', fontsize=20)

ax1=fig.add_subplot(221) # 2 rows 3 columns and the first axis
ax1.set_title('Price vs Total Sales Line Plot')
ax1.plot(car_sales['Price'], car_sales['Total Sales'])

ax2=fig.add_subplot(222)
ax2.set_title('Odometer vs Price Scatter Plot')
ax2.scatter(car_sales['Odometer (KM)'], car_sales['Price'])

ax3=fig.add_subplot(223)
ax3.set_title('Make vs Odometer Bar Plot')
ax3.bar(car_sales['Make'], car_sales['Odometer (KM)'])

ax3=fig.add_subplot(224)
ax3.set_title('Histogram of Car Make')
ax3.hist(car_sales['Make'])

plt.show()
```
